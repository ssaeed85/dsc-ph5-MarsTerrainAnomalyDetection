```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.image import imread
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Normalizing data for 8 bit
test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    'images/test/',
    target_size = (227,227),
    batch_size=8,
    color_mode='grayscale',
    class_mode='categorical'
)
```

```python
test_set_images = [test_generator[0][0][i] for i in range(0,5)]
```

```python
for each in test_set_images:
    matrix = each
    matrix = matrix[tf.newaxis, :, :, :]
    mp = tf.nn.max_pool2d(matrix, ksize=(5,5), strides=(5,5),
                              padding="SAME")
    ap = tf.nn.avg_pool2d(matrix, ksize=(5,5), strides=(5,5), padding="SAME")
    
    
    fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,5))

    ax1.imshow(matrix[0],cmap='gray')
    ax2.imshow(mp[0],cmap='gray')
    ax3.imshow(ap[0],cmap='gray')
    
    ax1.set_title('Original')
    ax2.set_title('MaxPool')
    ax3.set_title('AvgPool')
    
    plt.show()
```
